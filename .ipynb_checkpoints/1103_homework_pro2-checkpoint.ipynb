```python
class Kiosk:
    def __init__(self):
        self.menu  = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        # 누적 정보
        self.order_menu  = []
        self.order_price = []
        self.total = 0
        # 선택값들
        self.order_type = None   # 매장/포장
        self.temp = None         # HOT/ICE
        self.size = None         # SMALL/LARGE

    # 매장/포장 선택 (주문 전에 호출)
    def choose_place(self):
        while True:
            s = input("이용 방법을 선택하세요 (1: 매장, 2: 포장): ")
            if s == "1":
                self.order_type = "매장"
                print("→ 매장에서 먹음.\n")
                break
            elif s == "2":
                self.order_type = "포장"
                print("→ 포장해서 감.\n")
                break
            else:
                print("1 또는 2 중에서 선택해 주세요.")

    def menu_print(self):
        print("\n[ MENU ]")
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}원")

    # 메뉴 선택/온도/사이즈/누적 (0이면 종료)
    def menu_select(self):
        while True:
            self.menu_print()
            s = input("\n음료 번호를 입력하세요 (주문 완료 시 = 0): ")

            if s == "0":
                break
            if not s.isdigit():
                print("숫자를 입력해 주세요.")
                continue

            n = int(s)
            if not (1 <= n <= len(self.menu)):
                print("없는 메뉴입니다. 다시 선택해 주세요.")
                continue

            name  = self.menu[n-1]
            price = self.price[n-1]

            # 1) 온도 선택
            t = None
            while t not in (1, 2):
                try:
                    t = int(input("HOT은 1, ICE는 2를 입력하세요: "))
                except ValueError:
                    print("숫자로 입력해주세요.")
                    continue
                if t not in (1, 2):
                    print("1 또는 2 중에서 입력해주세요.")
            self.temp = "HOT" if t == 1 else "ICE"

            # 2) 사이즈 선택 (S/L)
            u = None
            while u not in (1, 2):
                try:
                    u = int(input("SMALL 사이즈는 1, LARGE 사이즈(+500원 추가)는 2를 선택하세요."))
                except ValueError:
                    print("숫자로 입력해주세요.")
                    continue
                if u not in (1, 2):
                    print("1 또는 2 중에서 선택해주세요.")
            self.size = "SMALL" if u == 1 else "LARGE"

            if self.size == "LARGE" : 
                price += 500 # 라지 선택시만 가격 추가

            # 누적
            label = f"{self.temp} {self.size} {name}"
            self.order_menu.append(label)
            self.order_price.append(price)
            self.total += price
            print(f"추가 메뉴: {label} : {price}원\n주문 금액 합계: {self.total}원")

    def pay(self):
        print(f"\n지불하실 금액은 {self.total}원 입니다.")
        while True:
            m = input("결제 수단을 선택하세요 (1: 현금, 2: 카드): ")
            if m == "1":
                print("직원을 호출하겠습니다.")
                break
            elif m == "2":
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("1 또는 2 중 하나를 입력해주세요.")

    def table(self):  # 주문서 출력 
        print('⟝' + '-' * 30 + '⟞')
        for _ in range(5):
            print('|' + ' ' * 31 + '|')

        # 매장/포장 정보
        if self.order_type:
            print(f"이용 방법 : {self.order_type}")

        # 주문 상품명 : 금액
        if not self.order_menu:
            print("주문 내역이 없습니다.")
        else:
            for i in range(len(self.order_menu)):
                print(f"{self.order_menu[i]} : {self.order_price[i]}원")
            print('합계 금액 :', sum(self.order_price), '원')

        for _ in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()
a.choose_place()   # 매장 or 포장
a.menu_select()    # 온도 → 사이즈(+500) → 누적 반복
a.pay()
a.table()
```
