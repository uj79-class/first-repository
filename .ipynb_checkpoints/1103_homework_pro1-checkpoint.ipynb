```python
#프로젝트 1
class Square:
    def __init__(self):
        self.square = int(input('도형 선택: 1.직사각형 2.평행사변형 3.사다리꼴 >>> '))
        if self.square == 1:
            self.rect()
        elif self.square == 2:
            self.par()
        elif self.square == 3:
            self.trape()
        else:
            print('잘못된 입력입니다.')

    def rect(self):
        w, v = map(int, input("가로,세로 입력 (예: 10,20) >>> ").split(','))
        print(f"직사각형의 넓이: {w * v}")

    def par(self):
        b, h = map(int, input("밑변, 높이 입력 (예: 10,20) >>> ").split(','))
        print(f"평행사변형의 넓이: {b * h}")

    def trape(self):
        b, u, h = map(int, input("밑변, 윗변, 높이 입력 (예: 10,20,30) >>> ").split(','))
        print(f"사다리꼴의 넓이: {(b + u) * h / 2}")
```

```python
 a = Square()
```
