```python
#은행 계좌 만들기 1. Account 클래스

#필요한 내중 모듈 부르기
import random  #계좌번호 생성 관련
from datetime import datetime  #계좌입출금 일자 생성 관련

#1.초기 설정(속성) 메서드 만들기 
class Account:
    #클래스 변수(모든 인스턴스가 공유) 만들기
    account_count = 0   #계좌 객체의 개수
    
    #인스턴스 변수(객체마다 독립) 만들기
    def __init__(self, name, balance): #입력받을 데이터() 
        self.name = name
        self.balance = balance
        self.bank = "SC은행"      # 고정 제공 
        self.account_num = self.generate_account_num() #함수로 만들어 제공
        self.deposit_count = 0  #입금 횟수 카운터 (이자지급 관련)
        self.deposit_log = []  #입금 내역 로그를 담을 빈 리스트 만들기
        self.withdraw_log = []  #출금 내역 로그를 담을 빈 리스트 만들기


        #객체 생성마다 계좌개수 +1
        Account.account_count += 1
        

        #계좌넘버 3-2-6자리로 랜덤 생성
    def generate_account_num(self):
        n1 = str(random.randint(0,999)).zfill(3) #999안에서 정수 랜덤 픽 후 자릿수 채움
        n2 = str(random.randint(1,99)).zfill(2)
        n3 = str(random.randint(0,999999)).zfill(6)
        account_num = f"{n1}-{n2}-{n3}"
        return account_num

    @classmethod  #3. 클래스 변수 출력: Account 클래스 실행에 의해 생성된 값
    def get_account_num(cls):               #클래스에 의한 것이므로 클래스 자신 cls 인자 
        print(f"생성된 계좌 수: {cls.account_count}")
    
    #4.입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print ("입금은 최소 1원 이상만 가능합니다.")
            return False
        else:
            self.balance += amount #잔액 업뎃
            self.deposit_count += 1 #입금횟수 업뎃

            #10. 새로운 입금 시간과 기록저장
            time = datetime.now().strftime("%Y-%m-%d %H:%M:%S") #datetime 문자열형태로변환
            self.deposit_log.append((time, amount)) #입금 내역 로그에 튜플로 시간과 금액추가
            
            print(f"입금 완료: {amount}원 / 남은 잔액 : {self.balance}원") #입금 출력 내역

            #5회마다 이자(잔액의 1%) 지급
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01) #소수점버림
                self.balance += interest
                print (f"연속 5회 입금 달성에 따른 이자 {interest:,}원 입금 완료")
                print (f"현재 잔액 : {self.balance:,}원")

            return True
    
    #5.출금 메서드
    def withdraw(self, amount):
        if self.balance < amount:
            print("잔액이 부족합니다.")
            return False
        else: 
            self.balance -= amount

            #새로운 출금 시간과 기록저장
            time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            self.withdraw_log.append((time, amount)) #로그를 튜플로 저장
            print(f"출금 완료: {amount}원 / 남은 잔액 : {self.balance}원")
            return True

    #6.정보출력 메서드
    def display_info(self):
        print(
            f"은행이름: {self.bank}, 예금주: {self.name}," 
            f"계좌번호: {self.account_num}, 잔고: {self.balance:,}"
            ) 
        #문자열 포맷지정자 콤마를 써서 1000단위 콤마 표시 

    # 입금 내역 기록코드 메서드
    def deposit_history(self):
        print(f"\n{self.name}의 입금 내역")
        for date, amount in self.deposit_log:  #입금 내역에서 날짜와 금액을 반복적으로 가져오기
            print(f"{date}: {amount}원")


    #출금 내역 기록코드 메서드
    def withdraw_history(self):
        print(f"\n{self.name}의 출금 내역")
        for date, amount in self.withdraw_log:
            print(f"{date}: {amount}원")
    
```

```python
#실행 코드 
#8. 클래스 Account의 인스턴스 생성 
a1 = Account("홍길동", 10000)
a2 = Account("김철수", 5000)
a3 = Account("박의진", 1000000)

print(a1.bank, a1.name, a1.account_num, a1.balance)
print(a2.bank, a2.name, a2.account_num, a2.balance)
print(a3.bank, a3.name, a3.account_num, a3.balance)
print("총 계좌 수:", Account.account_count)

#생성된 인스턴스는 고객 리스트에 저장 
customer = [a1, a2, a3]

#특정 고객 정보 출력
print(f"\n[잔고 100만원 이상 고객 명단]:")
for idx, i in enumerate(customer, start=1):
    if i.balance >= 1000000:
        print(f"{idx}. ", end="") # end=""는 출력 시 줄바꿈없이 이어서 오는 출력을 붙여라
        i.display_info()

print()


a1.deposit(100000)
a1.deposit(2000)
a1.withdraw(2000)
a1.withdraw(1000)

#입금 내역 출력
print()
print(a1.deposit_log)

#출금 내역 출력
print()
print(a1.withdraw_log)




```
